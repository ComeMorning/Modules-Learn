# API说明

## ApplicationMessage消息实例

```py
class hbmqtt.session.ApplicationMessage(packet_id, topic, qos, data, retain)
```
ApplicationMessage以及其子类用于存储已发布的消息信息流，这些对象根据其创建的方式（发送和接收）包含了不同的信息和QoS服务质量。

### 方法：build_publish_packet

根据ApplicationMessage的实例属性构建一个 `hbmqtt.mqtt.publish.PublishPacket` 包

```py
msg.build_publish_packet(dup=False)
```

参数：

- dup：强制导出标识。

返回：

`hbmqtt.mqtt.publish.PublishPacket`

### 属性：data

发布消息的载荷。

### 属性：packet_id

发布消息的[包标识符 id](http://docs.oasis-open.org/mqtt/mqtt/v3.1.1/os/mqtt-v3.1.1-os.html#_Toc398718025)

### 属性：puback_packet

`hbmqtt.mqtt.puback.PubackPacket` 实例对应的[PUBACK](http://docs.oasis-open.org/mqtt/mqtt/v3.1.1/os/mqtt-v3.1.1-os.html#_Toc398718043) 确认包。  
如果QoS != 1，或者 PUBCOMP 包还没发送或者没收到 的情况下 为`None`

### 属性：pubcomp_packet

`hbmqtt.mqtt.puback.PubrelPacket` 实例对应的 [PUBCOMP](http://docs.oasis-open.org/mqtt/mqtt/v3.1.1/os/mqtt-v3.1.1-os.html#_Toc398718058) 包  
如果 PUBLISH 包还没发送或者没收到的情况下为`None`

### 属性：pubrec_packet

`hbmqtt.mqtt.puback.PubrecPacket` 实例对应的 [PUBREC](http://docs.oasis-open.org/mqtt/mqtt/v3.1.1/os/mqtt-v3.1.1-os.html#_Toc398718048)包  
如果QoS != 2，或者 PUBREC 包还没发送或者没收到的情况下为`None`

### 属性：pubrel_packet

`hbmqtt.mqtt.puback.PubrelPacket`实例对应的 [PUBREL](http://docs.oasis-open.org/mqtt/mqtt/v3.1.1/os/mqtt-v3.1.1-os.html#_Toc398718053)  
如果 QoS != 2，或者 PUBREL 包还没发送或者还没收到的情况下为 `None`

### 属性：qos

消息服务质量

### 属性：retain

消息的retain值

### 属性：topic

消息 topic
