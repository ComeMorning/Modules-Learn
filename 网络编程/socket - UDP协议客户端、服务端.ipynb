# socket

## Server

```python
import time
import socket

local_addr = ('127.0.0.1', 10012)
buff_len = 1024

if __name__ == '__main__':

    # 创建一个socket
    udp = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
    # 绑定监听的地址和端口
    udp.bind(local_addr)

    while True:

        print('wating for message...')
        # 阻塞接收客户端数据
        data, addr = udp.recvfrom(buff_len)
        print('received from %s data:%s' % (addr, data.decode('utf-8')))

        time.sleep(1.5)
        # 发送数据给客户端
        udp.sendto(('server ' + str(time.time())).encode('utf-8'), addr)

    # 关闭socket
    udp.close()
```

## Client

```python
import time
import socket

server_addr = ('127.0.0.1', 10012)
buff_len = 1024

if __name__ == '__main__':

    # 创建一个socket
    udp = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)

    # 发送数据给服务器
    udp.sendto('hello server.'.encode('utf-8'), server_addr)
    # 阻塞接收服务器数据
    print(udp.recv(buff_len).decode('utf-8'))

    while True:
        udp.sendto(str(time.time()).encode('utf-8'), server_addr)
        
#         data, server_addr = client.recvfrom(1024)
        print(udp.recv(buff_len).decode('utf-8'))

    # 关闭socket
    udp.close()
```

## 服务端

```python
import socketserver


class MyHandler(socketserver.BaseRequestHandler):
    def handle(self):
        print('listening_in_handle')
        print(self.client_address)
        print(self.request)
        data, file = self.request
        print('客户消息', data)
        
        file.sendto(data.upper(), self.client_address)

local_addr = ('127.0.0.1', 10012)

if __name__ == '__main__':
    s = socketserver.ThreadingUDPServer(local_addr, MyHandler)
    # 循环建立连接，每建立一个连接就会启动一个线程（服务员）+调用Myhanlder类产生一个对象，
    s.serve_forever()
```
